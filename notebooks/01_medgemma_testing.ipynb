# MedGemma Tumor Board Prototype\n\nThis notebook tests MedGemma 1.5 through Ollama on synthetic oncology cases.

```python
import pandas as pd\nimport ollama\n\nMODEL = 'medgemma1.5'
```

```python
df = pd.read_csv('../data/synthetic_patients/sample_patients.csv')\ndf
```

```python
patient = df.iloc[0].to_dict()\npatient_data = '\\n'.join([f'{k}: {v}' for k, v in patient.items()])\n\nprompt = f'''\nYou are assisting a multidisciplinary oncology tumor board.\n\nUse ONLY the patient information provided.\nDo not invent missing biomarkers, imaging findings, pathology, staging, or treatment history.\n\nReturn:\n1. One-line case summary\n2. Key clinical facts\n3. Missing or unclear information\n4. MDT discussion questions\n5. Treatment considerations, phrased as possibilities, not final recommendations\n\nPatient data:\n{patient_data}\n'''\n\nresponse = ollama.chat(\n    model=MODEL,\n    messages=[{'role': 'user', 'content': prompt}]\n)\n\nprint(response['message']['content'])
```

## Batch test all patients

```python
outputs = []\n\nfor _, row in df.iterrows():\n    patient_data = '\\n'.join([f'{k}: {v}' for k, v in row.to_dict().items()])\n    response = ollama.chat(\n        model=MODEL,\n        messages=[{'role': 'user', 'content': prompt.replace(patient_data, patient_data)}]\n    )\n    outputs.append({'patient_id': row['patient_id'], 'summary': response['message']['content']})\n\noutputs[0]['summary']
```
